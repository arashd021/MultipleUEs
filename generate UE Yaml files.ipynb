```python
import os

# read the template file
with open('template.yaml', 'r') as f:
    template_lines = f.readlines()

# loop over range 100 to generate 100 files
for i in range(200):
    # increment the SUPI number by i
    sufi_line = [line for line in template_lines if line.startswith('supi:')][0]
    if 'imsi-' in sufi_line:
        sufi_num = str(int(sufi_line.split('-')[1].strip().strip("'")) + 1)
        print(sufi_num)
        template_lines = [line.replace(sufi_line, f"supi: imsi-{sufi_num}\n") if line.startswith('supi:') else line for line in template_lines]
    
    # create a new file with incremented SUPI number
    with open(f'UE_{i+4}.yaml', 'w') as f:
        f.writelines(template_lines) 
```
